#Q1: How has average energy consumption (electricity and gas) changed from 2004–2024?

```python
# Filter to municipality-level records only, to avoid double-counting neighborhoods
df_mun = df[df['recs'].isin(['Gemeente',"G"])]

# Display the first 5 rows
print(df_mun.head())
```

```python
# Group national averages per year
yearly_avg = df_mun.groupby('jaar').agg({
    'g_ele': 'mean',
    'g_gas': 'mean'
}).reset_index()

# Keep only rows where both g_ele and g_gas are not zero or NaN
yearly_avg_clean = yearly_avg[(yearly_avg['g_ele'] > 0) & (yearly_avg['g_gas'] > 0)]

print(yearly_avg_clean['jaar'])
```

```python
#print(yearly_avg_clean)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression

#Energy types
energy_types = ['g_ele', 'g_gas']
energy_types_avg = df_mun.groupby('jaar')[energy_types].mean().reset_index()

# Keep only rows where energy_types > 0
energy_types_avg_clean = energy_types_avg[(energy_types_avg ['g_ele']>0)&
                                          (energy_types_avg ['g_gas']>0)
]

# Prepare X values
X = energy_types_avg_clean['jaar'].values.reshape(-1, 1)
years = energy_types_avg_clean['jaar'].values

# Energy types and their labels with correct units
types = {
    'g_ele': ('Electricity', 'kWh'),
    'g_gas': ('Gas', 'm³')
}

plt.figure(figsize=(12,4))

colors = ['tab:blue', 'tab:orange']
i = 0

# Loop over energy type
for col, (label, unit) in types.items():
    y = energy_types_avg_clean[col].values

    # Line plot for actual yearly average values
    sns.lineplot(x=years, y=y, label=f'{label} Actual', linewidth=2, color=colors[i])

    # Linear regression
    model = LinearRegression()
    model.fit(X, y)
    y_pred = model.predict(X)

    # Plot regression line
    plt.plot(years, y_pred, linestyle='--', label=f'{label} Trend ({model.coef_[0]:.1f} {unit}/yr)', color=colors[i], linewidth=1)
    i += 1

# Plot formatting
plt.title('Average Energy Consumption and Trend Lines(2004–2023)')
plt.xlabel('Year')
plt.ylabel('Average Energy Consumption (kWh)')
plt.legend()
plt.grid(True)

# Set ticks only for the clean years
plt.xticks(energy_types_avg_clean['jaar'].unique(), rotation=45)

#plt.show()
plt.savefig('Q1_test.png', dpi=300, bbox_inches='tight')
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

X = yearly_avg_clean['jaar'].values.reshape(-1, 1)
y_elec = yearly_avg_clean['g_ele'].values

model_elec = LinearRegression()
model_elec.fit(X, y_elec)

print(f' Electricity Slope: {model_elec.coef_[0]:.2f} kWh/year')

y_gas = yearly_avg_clean['g_gas'].values
model_gas = LinearRegression()
model_gas.fit(X, y_gas)

print(f' Gas Slope: {model_gas.coef_[0]:.2f} m³/year')
```

The plot illustrates the national average household consumption of electricity and gas in the Netherlands from 2004 to 2023. Both electricity and gas consumption show a downward trend over time. Using linear regression, the analysis quantifies this decline.

# Q2: Which residential house types are more energy efficient?

### Electricity consumption per differen house type

```python
# Electricity Slopes for diffrerent building types

from sklearn.linear_model import LinearRegression
import numpy as np

# Electricity Slope appartement
X = type_avg_ele_clean['jaar'].values.reshape(-1, 1)
y_elec_ap = type_avg_ele_clean['g_ele_ap'].values

model_elec_ap = LinearRegression()
model_elec_ap.fit(X, y_elec_ap)

print(f'Electricity slope ap (apartment): {model_elec_ap.coef_[0]:.2f} kWh/year')

# Electricity Slope tussenwoning
y_elec_tw = type_avg_ele_clean['g_ele_tw'].values

model_elec_tw = LinearRegression()
model_elec_tw.fit(X, y_elec_tw)

print(f'Electricity slope hw(rowhouse): {model_elec_tw.coef_[0]:.2f} kWh/year')

# Electricity Slope hoekwoning
y_elec_hw = type_avg_ele_clean['g_ele_hw'].values

model_elec_hw = LinearRegression()
model_elec_hw.fit(X, y_elec_hw)

print(f'Electricity slope hw (corner house): {model_elec_hw.coef_[0]:.2f} kWh/year')

# Electricity Slope twee-onder-één-kap woning
y_ele_2w = type_avg_ele_clean['g_ele_2w'].values

model_ele_2w = LinearRegression()
model_ele_2w.fit(X, y_ele_2w)

print(f'Electricity slope 2w(semi-detached): {model_ele_2w.coef_[0]:.2f} kWh/year')

# Electricity Slope vrijstaande woning
y_ele_vw = type_avg_ele_clean['g_ele_vw'].values

model_ele_vw = LinearRegression()
model_ele_vw.fit(X, y_ele_vw)

print(f'Electricity slope vw(detached): {model_ele_vw.coef_[0]:.2f} kWh/year')

```

This visualization shows the average electricity consumption by housing type in the Netherlands from 2004 to 2023. All housing types exhibit a clear downward trend, with negative slopes indicating consistent reductions in electricity use over time. Detached houses remain the highest consumers, while apartments consume the least on average. However, apartments are also improving the slowest: their electricity use is decreasing by only about 16 kWh per year, compared to 43–44 kWh per year for rowhouses and corner houses. This suggests that while all housing types are becoming more energy efficient, larger improvements are occurring in low- and mid-density housing types.

### Trend analysis of gas consumption

```python
# Forecast of Gas Consumption Until 2035

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns

# Use Seaborn for cleaner default aesthetics
# sns.set(style="whitegrid", context="talk", palette="colorblind")

trend = type_avg_gas_clean.copy()
trend = trend[trend['jaar'] >= 2015]
trend = trend[trend['jaar'] != 2021]
trend = trend.set_index('jaar')
trend.loc[2021, :] = (trend.loc[2020,:] + trend.loc[2022,:]) / 2
trend = trend.sort_index()

types = {
    'g_gas_ap': 'Apartment',
    'g_gas_tw': 'Rowhouse',
    'g_gas_hw': 'Corner house',
    'g_gas_2w': 'Semi-detached house',
    'g_gas_vw': 'Detached house'
}

trend = trend.rename(columns=types)

colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:brown', 'tab:red']

# Prepare data
X = trend.index.values.reshape(-1, 1)
future_years = np.arange(trend.index.min(), 2036).reshape(-1, 1)
predictions = pd.DataFrame(index=range(trend.index.min(), 2036))

# Fit and predict
models = {}
for col in trend.columns:
    y = trend[col].values
    model = LinearRegression()
    model.fit(X, y)
    models[col] = model
    predictions[col] = model.predict(future_years)

# Plot
fig, ax = plt.subplots(figsize=(10, 5))

for i, col in enumerate(trend.columns):
    # Historical
    ax.plot(trend.index, trend[col], marker='o', linewidth=2, label=f'{col} Observed', color=colors[i])
    # Forecast
    ax.plot(predictions.index, predictions[col], linestyle='--', linewidth=1, label=f'{col} Forecast', color=colors[i])

# Visual clarity enhancements
ax.set_title('Forecast of Gas Consumption Until 2035', fontsize=18, weight='bold')
ax.set_xlabel('Year', fontsize=14)
plt.ylabel('Average Gas Consumption (m³)')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.grid(True, which='major', linestyle='--', linewidth=0.5, alpha=0.7)
ax.legend(frameon=False, loc='upper right')
plt.xticks(predictions.index,
           rotation=45
           )
plt.tight_layout()
#plt.show()
plt.savefig('Q2trend_test.png', dpi=300, bbox_inches='tight')

```

This code forecasts average gas consumption by housing type in the Netherlands through 2035, using linear regression models trained on data from 2015–2023 (with 2021 interpolated). It includes five residential types: apartments, rowhouses, corner houses, semi-detached, and detached homes. The model fits historical trends for each housing type and extends them into the future.

The forecast shows that although all housing types are reducing gas consumption, detached houses—historically the highest consumers—are declining at the fastest rate. Based on this trend, detached homes could approach the consumption levels of other types by around 2035.

However, this convergence is theoretical. In practice, differences in size, insulation, and heating needs mean that complete equalization is unlikely to occur, even if the trajectories suggest otherwise.

# Q3: How energy usage (electricity & gas) per person changes as household size increases?

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Copy clean yearly data
df_clean = df_mun[
    (df_mun['g_hhgro'] > 0) &
    (df_mun['g_ele'] > 0) &
    (df_mun['g_gas'] > 0)
].copy()

# Calculate per person energy use
df_clean['elec_per_person'] = df_clean['g_ele'] / df_clean['g_hhgro']
df_clean['gas_per_person'] = df_clean['g_gas'] / df_clean['g_hhgro']

# Group data by household size and average energy. Группирует по размеру домохозяйства от двух колонок, которые только создали. Беред среднее значение
grouped = df_clean.groupby('g_hhgro')[['elec_per_person', 'gas_per_person']].mean().reset_index()

# Plotting
plt.figure(figsize=(12,4))

sns.scatterplot(data=grouped, x='g_hhgro', y='elec_per_person', label='Electricity (kWh)', s=100)
sns.scatterplot(data=grouped, x='g_hhgro', y='gas_per_person', label='Gas (m³)', s=100)

# Linear trend lines
for col, label in [('elec_per_person', 'Electricity'), ('gas_per_person', 'Gas')]:
    X = grouped['g_hhgro'].values.reshape(-1,1)
    y = grouped[col].values
    model = LinearRegression()
    model.fit(X, y)
    y_pred = model.predict(X)
    plt.plot(grouped['g_hhgro'], y_pred, linestyle='--', label=f'{label} Trend ({model.coef_[0]:.0f} per person per extra HH member)')

# Labels and formatting
plt.title('Energy Consumption per Person vs. Household Size')
plt.xlabel('Average Household Size')
plt.ylabel('Energy Consumption per Person')
plt.grid(True)
plt.legend()
plt.savefig('Q3_hhsize.png', dpi=300, bbox_inches='tight')
# plt.show()
```

By analyzing both electricity and gas usage per person, the plot shows that larger households are more energy-efficient. In contrast, smaller households, such as those with one or two members, tend to consume more energy per individual.

# Q4: Are newer homes (post-2000) more energy-efficient than older homes?

```python
# How energy consumption change depending on the percenatge of the older(built before 2000) houses on the municipality level.
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Copy clean yearly data
df_clean2k = df_mun[
    (df_mun['pw_vo2k'] ) &
    #(df_mun['pw_na2k']> 0) &
    (df_mun['g_ele'] > 0) &
    (df_mun['g_gas'] > 0) &
    (df_mun['jaar'] > 2013) &
    (df_mun['jaar'] < 2023)
].copy()

# Group data by percentage of OLD houses
grouped_2k = df_clean2k.groupby('pw_vo2k')[['g_ele', 'g_gas']].mean().reset_index()

# Plotting
plt.figure(figsize=(10,4))

sns.scatterplot(data=grouped_2k, x='pw_vo2k', y='g_ele', label='Electricity (kWh)', s=100)
sns.scatterplot(data=grouped_2k, x='pw_vo2k', y='g_gas', label='Gas (m³)', s=100)

# Linear trend lines
for col, label in [('g_ele', 'Electricity'), ('g_gas', 'Gas')]:
    X = grouped_2k['pw_vo2k'].values.reshape(-1,1)
    y = grouped_2k[col].values
    model = LinearRegression()
    model.fit(X, y)
    y_pred = model.predict(X)
    plt.plot(grouped_2k['pw_vo2k'], y_pred, linestyle='--', label=f'{label} Trend ({model.coef_[0]:.0f} per house per extra 1% of houses built before 2000)')

# Labels and formatting
plt.title('Energy Consumption per House vs. Percentage of houses built before 2000')
plt.xlabel('Percentage of houses built before 2000')
plt.ylabel('Energy Consumption per House')
plt.grid(True)
plt.xticks(np.arange(50, 105, 5))  # X-ticks every 5%
plt.legend()
#plt.show()
plt.savefig('Q4_before2k.png', dpi=300, bbox_inches='tight')
```

# Q5: Do densely populated neighborhoods have different energy consumption patterns than sparse ones?

```python
# for all the years: looking at diferent neghbouhoods in The Hague with different population density
# Strip whitespace and filter the DataFrame for 's-Gravenhage' only
df_gravenhage = df[df['gm_naam'].str.strip() == "'s-Gravenhage"]

# Filter to neghbouhood-level in the hague
df_denhaag = df_gravenhage[(df_gravenhage['recs'].isin(['Buurt', 'B'])) & (df_gravenhage['gm_naam'] == "'s-Gravenhage")]

# Clean data: keep rows with valid density and energy values
df_clean_dhbdich = df_denhaag[
    (df_denhaag['bev_dich'] > 0) &
    (df_denhaag['g_ele'] > 0) &
    (df_denhaag['g_gas'] > 0)
].copy()

# Plotting
plt.figure(figsize=(12, 5))

sns.scatterplot(data=df_clean_dhbdich, x='bev_dich', y='g_ele', label='Electricity (kWh)', s=100)
sns.scatterplot(data=df_clean_dhbdich, x='bev_dich', y='g_gas', label='Gas (m³)', s=100)

# Linear regression lines
for col, label in [('g_ele', 'Electricity'), ('g_gas', 'Gas')]:
    X = df_clean_dhbdich['bev_dich'].values.reshape(-1, 1)
    y = df_clean_dhbdich[col].values
    model = LinearRegression()
    model.fit(X, y)
    y_pred = model.predict(X)
    plt.plot(df_clean_dhbdich['bev_dich'], y_pred, linestyle='--',
             label=f'{label} Trend (slope: {model.coef_[0]:.2f})')

# Formatting
plt.title('Energy Consumption vs. Population Density in The Hague')
plt.xlabel('Population Density (people per km²)')
plt.ylabel('Energy Consumption per House')
plt.grid(True)
plt.legend()
#plt.show()
plt.savefig('Q5_bev_dich_DH.png', dpi=300, bbox_inches='tight')
```

In The Hague, neighborhoods with higher population density tend to have lower average household energy consumption. This inverse relationship is consistent for both electricity and gas usage. The trend suggests that denser areas—often associated with smaller dwellings and shared infrastructure—are generally more energy-efficient on a per-household basis.
